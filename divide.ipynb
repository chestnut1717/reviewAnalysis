```python
import pandas as pd
```

```python
p1 = pd.read_excel("data/긍정_편의점샌드위치_존맛.xlsx")
p2 = pd.read_excel("data/긍정_편의점샌드위치_최고.xlsx")
n1 = pd.read_excel("data/부정_편의점샌드위치_맛없음.xlsx")
n2 = pd.read_excel("data/부정_편의점샌드위치_최악.xlsx")

tmp = [p1, p2, n1, n2]
```

```python
positive = pd.DataFrame({'Title' : [], 'Text': [], 'URL': [], 'INDEX': []})
negative = pd.DataFrame({'Title' : [], 'Text': [], 'URL': [], 'INDEX': []})
neutral = pd.DataFrame({'Title' : [], 'Text': [], 'URL': [], 'INDEX': []})
```

```python
for df in tmp:
    # 긍정
    positive_df = df[df["INDEX"] == 2].loc[:, ['Title', 'Text', "URL", "INDEX"]] 
    positive = pd.concat([positive, positive_df])
    
    # 부정
    negative_df = df[df["INDEX"] == 0].loc[:, ['Title', 'Text', "URL", "INDEX"]] 
    negative = pd.concat([negative, negative_df])
    
    # 중립
    neutral_df = df[df["INDEX"] == 1].loc[:, ['Title', 'Text', "URL", "INDEX"]]  
    neutral = pd.concat([neutral, neutral_df])

positive.reset_index(drop=True, inplace=True)
negative.reset_index(drop=True, inplace=True)
neutral.reset_index(drop=True, inplace=True)
```
